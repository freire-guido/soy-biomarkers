````python
import pandas as pd
````

````python
sra = pd.read_csv('../metadata/sra.csv', index_col=0)
biosamp = pd.read_csv('../metadata/biosample.csv', index_col=0)
````

````python
metadata = sra.join(biosamp, rsuffix='_biosamp')
metadata
````

# Anotaciones

````python
def matching_cols(row, pat):
    return row[row.astype(str).str.contains(pat, regex=False, case=False).fillna(False)].to_dict()
````

Junto palabras clave en kws. Guardo ```{columna: valor}``` en ```estreses``` si la columna contiene la palabra clave en su valor.

## Estreses

````python
kws = ['control', 'ctrl', 'stress', 'drought', 'bacteria', 'inoculated', 'flood', 'deficit', 'infest', 'pathogen', 'acid', 'fung', 'sporus', 'iron', 'aluminum', 'cadmium', 'cold', 'temperature', 'nitrogen', 'heat']
estreses = pd.DataFrame(index = metadata.index)

for kw in kws:
    estreses[kw] = metadata.apply(lambda r: matching_cols(r, kw), axis=1)
````

````python
ind_estr = estreses.map(lambda v: v != {})
ind_estr
````

````python
ind_estr.sum(axis=0)
````

````python
ind_estr.sum(axis=1)
````

````python
estreses = estreses[ind_estr.sum(axis=1) > 0]
ind_estr = ind_estr.loc[estreses.index]
estreses.to_csv('estreses.csv')
````

## Tejidos

````python
kws = ['leave', 'leaf', 'seed', 'crown', 'shoot', 'stem', 'hypocotyl', 'pod', 'root', 'embryo', 'cotyledon', 'petiole', 'radicle', 'root', 'ovule', 'ovary', 'embryo']
tejidos = pd.DataFrame(index = metadata.index)

for kw in kws:
    tejidos[kw] = metadata[['tissue_sam_ss_dpl145', 'tissue']].map(lambda c: str(c).replace('system', '')).apply(lambda r: matching_cols(r, kw), axis=1)
````

````python
ind_teji = tejidos.map(lambda v: v!= {})
ind_teji
````

````python
ind_teji.sum(axis=0)
````

````python
ind_teji.sum(axis=1)
````

````python
tejidos = tejidos[ind_teji.sum(axis=1) > 0]
ind_teji = ind_teji.loc[tejidos.index]
tejidos.to_csv('tejidos.csv')
````

# Condicion

````python
def cols_true(row):
# Devuelve lista de columnas verdaderas para row
    return row[row].index.to_list()
````

````python
condicion = pd.DataFrame(index=tejidos.index.union(estreses.index))
condicion
````

## Estrés

````python
ind_estr[ind_estr['control'] | ind_estr['ctrl']].index
````

````python
condicion.loc[ind_estr[ind_estr['control'] | ind_estr['ctrl']].index, 'estres'] = 'control'
````

````python
bioprojects = pd.read_csv('../metadata/venancio.csv', index_col=0)
bioprojects[bioprojects['stress'] == 'control']
````

````python
sample_stress_project = metadata.reset_index().set_index('bioproject').join(bioprojects, rsuffix='_bioprj').set_index('biosample')['stress']
condicion.loc[sample_stress_project == 'control', 'estres'] = 'control'
````

````python
control = condicion['estres'] == 'control'
````

````python
condicion.loc[~control, 'estres'] = ind_estr[~control].apply(cols_true, axis = 1)
condicion
````

Convierto listas de 1 elto. a str y (stress, [stress específico]) a [stress específico]

````python
lista = condicion['estres'].apply(lambda r: isinstance(r, list))

multi = condicion.loc[lista, 'estres'].map(len) > 1
condicion.loc[lista & multi, 'estres'] = condicion.loc[lista & multi, 'estres'].apply(lambda r: [s for s in r if s != 'stress'])

multi = condicion.loc[lista, 'estres'].map(len) > 1
condicion.loc[lista & ~multi, 'estres'] = condicion.loc[lista & ~multi, 'estres'].apply(lambda r: r[0])
condicion
````

Consolido los tipos de estres multiples

````python
condicion['estres'] = condicion.loc[condicion['estres'].notna(), 'estres'].apply(lambda r: str(set(r)) if type(r) == list else r)
condicion['estres'].unique()
````

````python
condicion['estres'] = condicion['estres'].str.replace("{'inoculated', 'nitrogen'}", "nutrient deficit")
condicion['estres'] = condicion['estres'].str.replace("{'pathogen', 'inoculated'}", "fungus")
condicion['estres'] = condicion['estres'].str.replace("{'deficit', 'heat'}", "heat")
condicion['estres'] = condicion['estres'].str.replace("{'deficit', 'cadmium', 'acid'}", "nutrient deficit")
condicion['estres'] = condicion['estres'].str.replace("{'deficit', 'acid', 'cadmium'}", "nutrient deficit")
condicion['estres'] = condicion['estres'].str.replace("{'deficit', 'acid'}", "nutrient deficit")
condicion['estres'] = condicion['estres'].str.replace("{'acid', 'deficit'}", "nutrient deficit")
condicion['estres'] = condicion['estres'].str.replace("{'nitrogen', 'inoculated'}", "nutrient deficit")
condicion['estres'] = condicion['estres'].str.replace("{'inoculated', 'pathogen'}", "bacteria")
condicion['estres'] = condicion['estres'].str.replace("{'heat', 'deficit'}", "nutrient deficit")
condicion['estres'] = condicion['estres'].str.replace("{'acid', 'nutrient deficit', 'deficit'}", "nutrient deficit")
condicion['estres'] = condicion['estres'].str.replace("{'cadmium', 'acid', 'deficit'}", "nutrient deficit")
condicion['estres'] = condicion['estres'].str.replace('temperature', 'heat')
condicion['estres'] = condicion['estres'].str.replace(r"^deficit", "nutrient deficit", regex=True)
condicion['estres'] = condicion['estres'].str.replace("nitrogen", "nutrient deficit")
condicion['estres'] = condicion['estres'].str.replace("cadmium", "nutrient deficit")
condicion['estres'] = condicion['estres'].str.replace("iron", "nutrient deficit")

condicion['estres'].value_counts()
````

````python
lista = condicion['tejido'].apply(lambda r: isinstance(r, list))

multi = condicion.loc[lista, 'tejido'].map(len) > 1
condicion.loc[lista & multi, 'tejido'] = condicion.loc[lista & multi, 'tejido'].apply(lambda r: [s for s in r if s != 'tejido'])

multi = condicion.loc[lista, 'tejido'].map(len) > 1
condicion.loc[lista & ~multi, 'tejido'] = condicion.loc[lista & ~multi, 'tejido'].apply(lambda r: r[0])
condicion
````

````python
condicion['tejido'] = condicion.loc[condicion['tejido'].notna(), 'tejido'].apply(lambda r: str(set(r)) if type(r) == list else r)
condicion['tejido'].unique()
````
