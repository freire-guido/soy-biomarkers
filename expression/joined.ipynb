```python
import pandas as pd
from glob import glob
```

```python
pd.read_csv('../data/expressions/exp_matrix_cotyledon_TPM_gene.tsv', delimiter='\t').transpose()
```

```python
tsvs = glob("../data/expressions/*.tsv")
expr = []
for tsv in tsvs:
    df = pd.read_csv(tsv, delimiter='\t').transpose()
    df.columns = df.iloc[0]
    df = df.drop('Gene')
    expr.append(df)
expr = pd.concat(expr)
expr
```

```python
expr.to_csv('joined.csv')
```

```python
expr.plot.density()
```
