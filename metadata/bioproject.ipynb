```python
import pandas as pd
import bioproject
```

```python
sra = pd.read_csv('sra.csv')
bioprojects = pd.DataFrame({'bioproject': sra['bioproject'].unique()})
bioprojects
```

```python
try:
    attr = pd.read_csv('bioproject_attributes.csv')
except:
    attr = bioprojects['bioproject'].apply(bioproject.get_bioproject)
    attr.index = bioprojects['bioproject']
    attr = attr.apply(bioproject.from_bioproject)
    attr.to_csv('bioproject_attributes.csv')
attr
```

```python
bioprojects = pd.DataFrame.from_records(attr, index=attr.index)
bioprojects
```

```python
bioprojects['abstract'] = bioprojects['doi'].apply(bioproject.get_abstract)
bioprojects
```

```python
bioprojects.to_csv('bioproject.csv')
```
