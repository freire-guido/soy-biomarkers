````python
import pandas as pd
````

````python
sra = pd.read_csv('../data/glycine_max_sra.csv')
sra
````

````python
sra = sra[sra['assay_type'] == 'RNA-Seq'].set_index('biosample')
sra
````

Capturamos información a nivel de muestra pero sra es a nivel de "corrida". Algunas muestras pueden tener más de una corrida:

````python
sra.index.duplicated().sum()
````

````python
sra = sra.loc[~sra.index.duplicated(keep='first')]
sra.index.duplicated().sum()
````

# attributes

Convertir columna "attributes" con formato [{k=<key>, v=<value>}] a columna "jattr" con formato json. Después uso ```pd.DataFrame.from_records``` para armar columnas para cada key.

````python
import jattr
````

````python
attr = sra['attributes'].apply(jattr.from_attributes)
attributes = pd.DataFrame.from_records(attr, index=attr.index)
attributes
````

````python
attributes.to_csv('sra_attributes.csv')
sra.drop('attributes', axis=1).to_csv('sra.csv')
````
