```python
import pandas as pd
import biosample
```

```python
sra = pd.read_csv('sra.csv')
sra
```

```python
try:
    attr = pd.read_csv('biosample_attributes.csv', index_col=0)
    attr['attr'] = attr['attr'].apply(eval)
except:
    attr = sra['biosample'].apply(biosample.get_biosample)
    attr.index = sra['biosample']
    attr['attr'] = attr.apply(biosample.from_biosample)
    attr.to_csv('biosample_attributes.csv')
attr
```

```python
biosample = pd.DataFrame.from_records(attr['attr'], index=sra['biosample'])
biosample
```

```python
attributes = pd.DataFrame.from_records(biosample['Attributes'].dropna(), index=biosample['Attributes'].dropna().index)
identifiers = pd.DataFrame.from_records(biosample['Identifiers'].dropna(), index=biosample['Identifiers'].dropna().index)
```

```python
biosample = pd.DataFrame(biosample['title']).join(attributes, rsuffix='_attr')
biosample = biosample.join(identifiers, rsuffix="_ident")
biosample
```

```python
biosample.to_csv('biosample.csv')
```
