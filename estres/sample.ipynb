````python
import pandas as pd
````

````python
metadata = pd.read_csv('../data/joined.csv', index_col=0)
metadata
````

````python
def matching_cols(row, pat):
    return row[row.astype(str).str.contains(pat, regex=False, case=False).fillna(False)].to_dict()
````

Junto palabras clave en kws. Guardo ```{columna: valor}``` en ```estreses``` si la columna contiene la palabra clave en su valor.

````python
kws = ['control', 'ctrl', 'stress', 'drought', 'bacteria', 'inoculated', 'flood', 'deficit', 'infest', 'pathogen', 'acid', 'fung', 'sporus']
estreses = pd.DataFrame(index = metadata.index)

for kw in kws:
    estreses[kw] = metadata.apply(lambda r: matching_cols(r, kw), axis=1)
````

````python
indicadores = estreses.map(lambda v: v != {})
indicadores
````

````python
indicadores.sum(axis=0)
````

````python
indicadores.sum(axis=1)
````

````python
estreses = estreses[indicadores.sum(axis=1) > 0]
indicadores = indicadores.loc[estreses.index]
estreses.to_csv('estreses.csv')
````

# Condición
## Estres vs. control

````python
def cols_true(row):
# Devuelve lista de columnas verdaderas para row
    return row[row].index.to_list()
````

````python
condicion = pd.DataFrame(index = estreses.index)

condicion.loc[indicadores['control'] | indicadores['ctrl'], 'condicion'] = 'control'
control = condicion['condicion'] == 'control'

condicion.loc[~control, 'condicion'] = indicadores[~control].apply(cols_true, axis = 1)
condicion
````

````python
multi = ~control & (condicion.loc[~control, 'condicion'].map(len) > 1)
condicion[multi]
````

Muchas muestras tienen stress además del estrés específico. Me quedo con este último.

````python
condicion.loc[multi, 'condicion'] = condicion.loc[multi, 'condicion'].apply(lambda r: [s for s in r if s != 'stress'], axis=1)
condicion[multi]
````

Repito a ver cuales quedaron con varios estreses

````python
multi = ~control & (condicion.loc[~control, 'condicion'].map(len) > 1)
condicion[multi]
````

````python
condicion.loc[multi, 'condicion'] = condicion.loc[multi, 'condicion'].apply(lambda r: [s for s in r if s != 'deficit' and s != 'inoculated'], axis=1)
condicion[multi]
````
