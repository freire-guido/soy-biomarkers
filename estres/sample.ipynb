````python
import pandas as pd
````

````python
sra = pd.read_csv('../metadata/sra.csv', index_col=0)
biosamp = pd.read_csv('../metadata/biosample.csv', index_col=0)
````

````python
metadata = sra.join(biosamp, rsuffix='_biosamp')
metadata
````

````python
def matching_cols(row, pat):
    return row[row.astype(str).str.contains(pat, regex=False, case=False).fillna(False)].to_dict()
````

Junto palabras clave en kws. Guardo ```{columna: valor}``` en ```estreses``` si la columna contiene la palabra clave en su valor.

````python
kws = ['control', 'ctrl', 'stress', 'drought', 'bacteria', 'inoculated', 'flood', 'deficit', 'infest', 'pathogen', 'acid', 'fung', 'sporus', 'iron', 'aluminum', 'cadmium', 'cold', 'temperature', 'nitrogen']
estreses = pd.DataFrame(index = metadata.index)

for kw in kws:
    estreses[kw] = metadata.apply(lambda r: matching_cols(r, kw), axis=1)
````

````python
indicadores = estreses.map(lambda v: v != {})
indicadores
````

````python
indicadores.sum(axis=0)
````

````python
indicadores.sum(axis=1)
````

````python
estreses = estreses[indicadores.sum(axis=1) > 0]
indicadores = indicadores.loc[estreses.index]
estreses.to_csv('estreses.csv')
````

# Condición
## Estres vs. control

````python
def cols_true(row):
# Devuelve lista de columnas verdaderas para row
    return row[row].index.to_list()
````

## Control

````python
condicion = pd.DataFrame(index = estreses.index)
condicion.loc[indicadores['control'] | indicadores['ctrl'], 'condicion'] = 'control'
````

````python
bioprojects = pd.read_csv('../metadata/venancio.csv', index_col=0)
bioprojects[bioprojects['stress'] == 'control']
````

````python
sample_stress_project = metadata.reset_index().set_index('bioproject').join(bioprojects, rsuffix='_bioprj').set_index('biosample')['stress']
condicion.loc[sample_stress_project == 'control', 'condicion'] = 'control'
````

````python
control = condicion['condicion'] == 'control'
````

## Estrés

````python
condicion.loc[~control, 'condicion'] = indicadores[~control].apply(cols_true, axis = 1)
condicion
````

````python
multi = ~control & (condicion.loc[~control, 'condicion'].map(len) > 1)
condicion[multi]
````

Muchas muestras tienen stress además del estrés específico. Me quedo con este último.

````python
condicion.loc[multi, 'condicion'] = condicion.loc[multi, 'condicion'].apply(lambda r: [s for s in r if s != 'stress'])
condicion[multi]
````

Repito a ver cuales quedaron con varios estreses

````python
multi = ~control & (condicion.loc[~control, 'condicion'].map(len) > 1)
condicion[multi]
````

````python
condicion.loc[multi, 'condicion'] = condicion.loc[multi, 'condicion'].apply(lambda r: [s for s in r if s != 'deficit' and s != 'inoculated'])
condicion[multi]
````

Ahora los transformo a ```str```

````python
lista = condicion['condicion'].apply(lambda r: isinstance(r, list))
lista
````

````python
condicion.loc[lista, 'condicion'] = condicion.loc[lista, 'condicion'].apply(lambda r: r[0])
````
